```python
#New Dataframe with only columns with numbers
Annual_Average_Unemployment1 = Annual_Average_Unemployment[['State', '2017 Rate', '2018 Rate', '2020 Rate','2021 Rate']]
Annual_Average_Unemployment1.head()
```

# Correlation

```python
Annual_Average_Unemployment1.corr()
```

# Seaborn Correlation Heatmap

```python
plt.figure(figsize=(8,8))
seaborn.heatmap(Annual_Average_Unemployment1.corr(), annot=True, cmap='coolwarm')
```

# Matplotlib Correlation Heatmap

```python
plt.figure(figsize=(16,8))
sns.heatmap(Annual_Average_Unemployment1.corr(), annot=True, annot_kws={"size": 14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

#  Bar Chart

```python
#Top 10 states with the Highest Unemployment Rate

top_10_highest_States_Unemployment_Averages = Annual_Average_Unemployment1.groupby(by = 'State').max()[['2017 Rate', '2018 Rate']].sort_values(by = ['State'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,8))
plt.title("Top 10 States with the Highest Unemployment Rate", size = 14)
ax = sns.barplot(data = top_10_highest_States_Unemployment_Averages.iloc[:10], y = "2017 Rate", x = "State", linewidth = 2, edgecolor = 'red')
plt.xlabel("State")
plt.ylabel("2017 Rate")
plt.show()
```
