# Matplotlib Correlation Heatmap

```python
plt.figure(figsize=(16,8))
sns.heatmap(Feeding_People1.corr(), annot=True, annot_kws={"size": 14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
